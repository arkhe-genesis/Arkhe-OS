## Módulo 1: Instalação e primeiros passos

### Verificar instalação
Execute a célula abaixo para carregar a extensão Arkhe:

```python
%load_ext arkhe_ipython
```

### Verificar status do Safe Core

```python
%arkhe status
```

## Módulo 2: Escaneamento de código com %arkhe scan

Escaneie trechos de código para detectar vulnerabilidades:

```python
# Código seguro para teste
safe_code = '''
def hello(name: str) -> str:
    return f"Hello, {name}!"
'''
%arkhe scan $safe_code
```

## Módulo 3: Execução segura com %%arkhe secure

Execute células com proteção completa do Safe Core:

```python
%%arkhe secure
# Este código será verificado pelo Guardião Atratora antes da execução
import hashlib
print(f"🔐 Hash seguro: {hashlib.sha3_256(b'Arkhe').hexdigest()[:16]}")
```

## Módulo 4: Auditoria e rastreabilidade

Consulte registros de auditoria por selo temporal:

```python
# Obter selo da última execução
from arkhe_ipython import get_temporal_anchor
last_seal = get_temporal_anchor()
print(f"Último selo: {last_seal}")

# Consultar auditoria
%arkhe audit $last_seal
```

## Módulo 5: Perfis de domínio e campo atratora

Altere o perfil do campo atrator para diferentes tipos de tarefa:

```python
# Ver perfis disponíveis
%arkhe profile

# Mudar para perfil técnico (mais conservador)
%arkhe profile technical

# Verificar Φ_C atual
%arkhe phi-c
```

## Módulo 6: Conformidade MA‑S2

Verifique conformidade com o padrão MA‑S2:

```python
# Status geral de conformidade
%arkhe compliance

# Status por domínio específico
%arkhe compliance cvs
```

## Módulo 7: Auto‑remediação e integração GRC

Orquestre correções e sincronize com ferramentas de GRC:

```python
# Simular deploy de patch para CVE
%arkhe deploy CVE-2026-12345

# Sincronizar com ServiceNow/RSA Archer
%arkhe grc-sync CVE-2026-12345
```
